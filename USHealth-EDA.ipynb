```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
data = pd.read_csv("/kaggle/input/ushealth/UShealth.csv")
```

```python
smoker_charges = data.groupby('smoker')['charges'].mean()
region_charges = data.groupby('region')['charges'].mean()
```

# Exploratory Data Analysis (EDA)

## 1. Distribution of charges

```python
plt.figure(figsize=(8, 5))
plt.hist(data['charges'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Medical Charges')
plt.xlabel('Charges')
plt.ylabel('Frequency')
plt.show()
```

## 2. Distribution of BMI

```python
plt.figure(figsize=(8, 5))
plt.hist(data['bmi'], bins=30, color='lightgreen', edgecolor='black')
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

## 3. Average charges by smoking status (bar chart)

```python
smoker_charges.plot(kind='bar', color=['orange', 'blue'], figsize=(8,5))
plt.title('Average Medical Charges by Smoking Status')
plt.ylabel('Average Charges')
plt.xlabel('Smoking Status')
plt.show()
```

## 4. Average charges by region (bar chart)

```python
region_charges.plot(kind='bar', color='purple', figsize=(8,5))
plt.title('Average Medical Charges by Region')
plt.ylabel('Average Charges')
plt.xlabel('Region')
plt.show()
```

## 5. Scatter plot between BMI and charges

```python
plt.figure(figsize=(8, 5))
plt.scatter(data['bmi'], data['charges'], color='red')
plt.title('BMI vs Medical Charges')
plt.xlabel('BMI')
plt.ylabel('Medical Charges')
plt.show()
```

## 6. Scatter plot between Age and charges

```python
plt.figure(figsize=(8, 5))
plt.scatter(data['age'], data['charges'], color='green')
plt.title('Age vs Medical Charges')
plt.xlabel('Age')
plt.ylabel('Medical Charges')
plt.show()
```

## Matrica korelacije:

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
corr_matrix = data.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title('Correlation Matrix of the Dataset')
plt.show()
```

## Pozitivna korelacija BMI-a i starosti ukazuje na očekivan porast medicinskih troškova sa povečanjem istih vrijednosti
